```python
import pandas as pd
import seaborn as sns
import seaborn.objects as so
```

```python
df = pd.read_csv('./speed_benchmark_detect.tsv', sep='\t')
df = df.query("includes_loading == True")
```

```python

(
    so.Plot(df, x='batch_size', y='duration', color='model_cls')
    .facet(col="device")
    .share(y=True, x=True)
    .add(so.Dot(), so.Dodge(), so.Jitter(.3))
    .add(so.Line(), so.Agg('median'), so.Dodge(), so.Jitter(.3))
    .scale(
        x=so.Continuous(trans="log2")
        #y=so.Continuous(trans='log')
    )
    .label(
        x="Batch size", y="Duration (sec.)"
    )
)
```
